Encapsulation:
Encapsulation binds together data and functions that amnipulate the data, keeping it safe from interference and misuse.Encapsulation is one of the fundamental concepts in object-oriented programming(OOP).it describes the idea of wrapping data and the methods that work on data within one unit. This puts restrictions on accessing variables and methods directly and can prevent the accidental modification of data. To prevent accidental change, an object variable can only be changed by an objects method.Those types of variables are known as private variables.
**self. PUBLIC

**self._  PROTECTED

**self.__  PRIVATE

self.name= public

self._name= protected/private

self.__name= strongly private

```python
class Encapsulation():
  def __init__(self,a,b,c):
    self.a=a
    self._b=b
    self.__c=c

  def mymethod(self):
    print(self.__c)
x= Encapsulation(11,15,17)
x.mymethod()

```
