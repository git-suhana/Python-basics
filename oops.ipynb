SELF:
self represents the instance of a class. By using the 'self' we can access the attributes and methods of a class in python.
self is a parameter in function and user can use another parameter name in place of it.But it is advisable to use self as it increase the readability of code.


```python
class Person:
  def read(self):
    print('reading')
  def walk(self):
    print('walking')
obj= Person()
obj.read()
obj.walk()

obj1= Person()
obj1.read()
obj1.walk()
```
