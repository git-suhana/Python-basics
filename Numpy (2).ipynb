2D Slicing:

```python
import numpy as np
a=np.array([[1,2,'3',4,7],[14,32,'30',40,77]])
print(a.ndim)
print(a[0:2,2:5]) #0th and 1st row and 2nd 3rd 4th columns
print(a[:,1:3])   #0th and 1st row  and 1st 2nd columns
```

3D Slicing:

```python
import numpy as np
a=np.array([[[1,2,'3',4,7],[14,32,'30',40,77]],[[19,72,'38',94,97],[15,392,'320',400,770]]],dtype=int)
print(a)
print(a.ndim)
print(a[:,1,2:5])
print(a[:,:,1:4])
```

Array joining:

```python
import numpy as np
a= np.array([[[3,4,5],[2,3,4]],[[5,6,7],[9,8,7]]])
b=np.array([[[2,4,6],[1,3,5]],[[6,8,10],[7,9,1]]])
c=np.vstack((a,b))
print(c)
```

```python
import numpy as np
a= np.array([[2,3,4],[4,5,6]])
b=np.array([[8,9,7],[5,3,7]])
c= np.row_stack((a,b))
print(c)
```

```python
import numpy as np
a=np.array([[2,3,4],[20,30,40]])
b= np.array([[5,6,7],[50,60,70]])
c= np.hstack((a,b))
print(c)
```

```python
d= np.column_stack((a,b))
print(d)
```

```python
import numpy as np
a=np.array([[2,3,4],[9,8,7]])
b=np.array([[45,87,21],[67,44,45]])
c= np.concatenate((a,b))
print(c)
```

```python
import numpy as np
a=np.array([[2,3,4],[6,7,8]])
b=np.array([[4,5,6],[1,2,3]])
c= np.concatenate((a,b),axis=1)
print(c)
```
