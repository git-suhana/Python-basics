String Operations:
find, rfind, split

```python
s1= 'hello world'
x= s1.find('world')
print(x)
```

```python
x= s1.find('world',1,11)
print(x)
```

```python
x= s1.find('world',1, 7)
print(x)
```

```python
x= s1.rfind('world')
print(x)
```

```python
x= s1.split(" ")
print(x)
```

The isalnum() method returns true if all the characters are alphanumeric, meaning alphabet(a-z) and numbers(0-9).Example of characters that are not alphanumeric:(space)!#$%&.

```python
txt = "Company12"
x = txt.isalnum()
print(x)
```

```python
txt = "company 12"
x= txt.isalnum()
print(x)
```

Stringjoin:
to join the words to a string.

```python
s1= ['hello', 'world']
x= ' '.join(s1)
print(x)
```

Capitalize:
it will convert the fist letter of first word into capital.

```python
y= x.capitalize()
print(y)
```

Upper:
it will convert all letters into uppercase.

```python
y= x.upper()
print(y)
```

Lower:
it will convert all letters into lowercase.


```python
y=x.lower()
print(y)
```

Title:
it will convert the first letter of all the words into capital.

```python
y=x.title()
print(y)
```

istitle:
to check whether it satisfies the title rule.

```python
s1 = 'Hello World'
x= s1.istitle()
print(x)
```
