Pandas:
pandas is a python library used for working with data sets. it has functions for analyzing, cleaning, exploring and manipulating data. Pandas allows us to analyze big and make conclusions based on statistical theories. Pandas can clean messy data sets, and make them readable and relevant. Relevant data is very important in data science. in pandas, one dimension is called series. two dimesion is called data frame. three dimension is called panel.

```python
import pandas as pd
data = [3,2,5,8]
s= pd.Series(data)
print(s)
```

```python
import pandas as pd
import numpy as np
data1={'name':['shinas','adith','sarang'],'chem':[40,60,40],'maths':[40,60,40],'phy':[40,60,40]}

s=pd.DataFrame(data1)
s['bio']=[60,48,50]
s['total']=s['chem']+s['maths']+s['phy']+s['bio']
list1=[]
for i in s['total']:
    if i<=100:
      list1.append('blw avg')
    elif i<=200:
      list1.append('avg')
    else:
      list1.append('exclnt')
#print(list1)
s['status']=list1
print(s)
del s['chem']
s= s.drop('phy',axis=1)  #want to specify axis
s.pop('maths')
print(s)
print(s.ndim)
print(s.shape)
print(s.size)
print(s.columns)
print(s['status'].unique())
print(s['bio'].unique())
print(s['status'].value_counts())
print(s)
```
