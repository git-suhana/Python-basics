Data Visualization:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

glucose=df['Glucose'].to_list()
plt.hist([glucose],bins=5 ,rwidth=0.6,color= 'red')
plt.xlabel('glucose')
plt.ylabel('frequency')
plt.title('glucose histogram')
plt.show()
```

This histogram represents the frequency distribution of glucose measurements in this sample of data. The glucose level appears to be normal with the highest frequency concentrated between 85 to 120 mg. A few outliers are present at the lower end of glucose level indicating rare instances.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

age=df['Age'].to_list()
plt.hist([age],bins=5 ,rwidth=0.4,color= 'purple')
plt.xlabel('age')
plt.ylabel('Frequency')
plt.title('age histogram')
plt.show()
```

This histogram represents the different age groups in x axis plotted with its frequency in y axis. The data shows a peak around the age of 20-30 group, with a gradual decrease  in the frequency as age increases to 60.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

BMI=df['BMI'].to_list()
plt.hist([BMI],bins=5, rwidth=.6, color='green')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.title('BMI histogram')
plt.show()
```

This histogram represents the BMI in x axis and its mostly concentrated between a range of 25-35. A rise in BMI can directly affect the rate of diabetes as they are closely linked.

Boxplots:
To show the outliers, As the glucose and insulin may vary from the normal distributions, it may be represented using a box plot.

```python
from matplotlib import pyplot as plt
import seaborn as sns
sns.boxplot(y=df['Glucose'])
```

There are a few outliers at the lower end of glucose level causing rare instances.

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(y=df['Insulin'])
```

A large number of outliers are present at the higher end of insulin level. It indicates some of the exceptional cases of individuals whose insulin rate is at its peak.

```python
from matplotlib import pyplot as plt
glucose= df['Glucose'].to_list()
insulin= df['Insulin'].to_list()
plt.boxplot([glucose,insulin],showmeans=True,patch_artist=True)
plt.title('glucose and insulin boxplot')
plt.show()
```

This shows a comparison between the plots of glucose and insulin in a single one. The first one shows the boxplot of glucose, in which both the mean and median are close to each other . While in the second one showing the boxplot of insulin, the mean is pulled towards the higher frequency, ie, mean is greater than the median.

Heat Map

```python
numeric = df.select_dtypes(include='number')
correlation_matrix=numeric.corr()
sns.heatmap(correlation_matrix,annot=True)
plt.show()
```

This shows how the numeric data in the given sample is correlated with each other. Here the Pregnancy,Age and glucose are the factors that shows more correlation with the outcome.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
selected_colm=df[['Glucose','BMI','Age','Insulin','Outcome']]
correlation_matrix=selected_colm.corr()
sns.heatmap(correlation_matrix,annot=True)
plt.show()

```

This shows the heat map for the selected columns in the data, a change in these data can easily affect the outcome.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(x='Glucose',y='BMI',data=df,hue='Outcome')
plt.title('Glucose vs BMI')
plt.show()
```

This shows the scatter plot between glucose and BMI.Mostly the affected and non affected people are lying in a range around 80-200 mg in glucose and 25-50 in BMI.As it is overlapped each other, it can be determined that many factors other than this are also playing an important role in balancing the outcome.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(x='Insulin',y='Age',data=df,hue='Outcome')
plt.title('Insulin vs Age')
plt.show()
```

This shows scatter plot between age and insulin. As the plot itself shows that most of the people are having insulin at the extreme lowest spectrum and it also varies to higher level irrespective of age.
