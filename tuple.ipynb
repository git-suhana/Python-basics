tuple:
its an ordered sequence , immutable, infinite length, iterable, allows duplicate members and return with () brackets.

```python
tuple = ("achu","kichu","lechu",10,4,7)
print(tuple)
```

functions in tuple:
count, sum, min, max, index

```python
tuple =(1,2,3,4,5)
print(tuple)
print (sum(tuple))
print(min(tuple))
print(max(tuple))
print(tuple.count(4))
print(tuple.index(4))
```
