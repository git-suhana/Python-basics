Polymorphism:
The word polymorphism means having many forms.Polymorphism means same function name being uses for different types. 2 types of polymorphism:
1)overloading- morethan one method name in a single class.
2)overriding

Function overloading not support in python: class having more than one function with same name.


```python
#overloading
class new:
  def hello(self,s,b):
    self.s=s
    self.b=b
    print(self.s+self.b)

  def hello(self,a,b,c):
    self.a=a
    self.b=b
    self.c=c
    print(self.a+self.b+self.c)

obj=new()
obj.hello(1,2,3)
obj.hello(1,2)
```

```python
class student:
  def sum(self,a,b):
    self.a=a
    self.b=b

  def sum(self,x=None,y=None,z=None):
    self.x=x
    self.y=y
    self.z=z
    s=0
    if x!=None and y!=None and z!=None:
      s=x+y+z
    elif x!=None and y!=None:
      s=x+y
    else:
      s=x
    return s

obj=student()
obj.sum(1,2)
```

```python
class point:
  def __init__(self,x,y):
    self.x=x
    self.y=y

  def __add__(self,other):
    x=self.x+other.x
    y=self.y+other.y
    return point(x,y)

p1= point(1,2)
p2= point(3,4)
print(p1+p2)
```

```python
class point:
  def __init__(self,x,y):
    self.x=x
    self.y=y

  def __add__(self,other):
    x=self.x+other.x
    y=self.y+other.y
    return point(x,y)
  def __str__(self):
    return f"{self.x},{self.y}"

p1= point(1,2)
p2= point(3,4)
print(p1+p2)
```
