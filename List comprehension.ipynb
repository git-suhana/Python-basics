List comprehension offer a shorter syntax when you want to create a new list based on the values of an existing list.

```python
list1=range(10,50)
even=[]
for i in list1:
    if i%2==0:
        even.append(i)
print(even)
```

```python
list1 = range(10, 50)
even2 = [i for i in list1 if i % 2 == 0]
print(even2)
```
