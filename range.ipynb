range:
to print a range of values



```python
x= range(0,10)
print(x)
for i in x:
  print(i)
```

step value:
the values to be incremented with a value

```python
x= range(3,20,3)
for i in x:
  print(i)
```

```python
x =  range(96,71,-4)
for i in x:
  print(i)
```
