uppercase letters in the string.

```python
x= str(input("enter a string"))
count=0
for i in x:
  if i.isupper():
    count+=1
print(count)
```

9. Create a program that checks if a given string is a palindrome.

```python
x = str(input("enter a string"))
y = x[::-1]
if x==y:
  print("palindrome")
else:
  print("not palindrome")
```
