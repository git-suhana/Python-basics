union:
returns a new set with all items from the both sets.


```python
n= {1,2,3,4,5}
m= {4,5,6,7,8}
print(n.union(m))
```

```python
n= {'sin', 'lin', 'bin'}
m= {'don','lon','mon'}
o= set.union(n,m)
print(o)
```

difference:
return a new set with only items from first set and not present in second.


```python
n= {'sin', 'lin', 'bin'}
m= {'don','lon','mon'}
p= set.difference(n,m)
print(p)
```

intersection:
returns the items that are presnet in both


```python
n= {'sin', 'lin', 'bin'}
m= {'don','lon','mon'}
q= set.intersection(n,m)
print(q)
```

isdisjoint:
returns whether sets have a intersection or not


```python
n= {'sin', 'lin', 'bin'}
m= {'don','lon','mon'}
r= set.isdisjoint(n,m)
print(r)
```

frozenset
immutasble version of set


```python
n= {'sin', 'lin', 'bin'}
m= frozenset(n)
print(m)
```
