To print even,odd numbers in a list between 100 and 500.

```python
x= range(100,501)
for i in x:
  odd= []
  even= []
  num= []
for i in x:
  num.append(i)
  if i%2==0:
    even.append(i)
  else:
    odd.append(i)

print(num)
print(odd)
print(even)
```

To print patterns:

```python
for i in range(1,6):
  for i in range(1,i+1):
    print(i,end=' ')
  print('\n')
```

To print backward :

```python
for i in range(1,6):
  for i in range(i,0,-1):
    print(i,end=' ')

  print('\n')
```

```python
dict1={'a':'vinu','b':234,'c':'hello'}
list1=[]
for i in dict1:
  list1.append(dict1.get(i))
print(list1)
```

to print both keys and values at a time:

```python
dict1={'a':'vinu','b':234,'c':'hello'}
list1=[]
list2=[]
for i,j in dict1.items():
  list1.append(i)
  list2.append(j)
print(list1)
print(list2)
```
