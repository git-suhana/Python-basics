Polymorphism:
The word polymorphism means having many forms.Polymorphism means same function name being uses for different types. 2 types of polymorphism:
1)overloading- morethan one method name in a single class.
2)overriding

Function overloading not support in python: class having more than one function with same name.


```python
#overloading
class new:
  def hello(self,s,b):
    self.s=s
    self.b=b
    print(self.s+self.b)

  def hello(self,a,b,c):
    self.a=a
    self.b=b
    self.c=c
    print(self.a+self.b+self.c)

obj=new()
obj.hello(1,2,3)
obj.hello(1,2)
```

```python
class student:
  def sum(self,a,b):
    self.a=a
    self.b=b

  def sum(self,x=None,y=None,z=None):
    self.x=x
    self.y=y
    self.z=z
    s=0
    if x!=None and y!=None and z!=None:
      s=x+y+z
    elif x!=None and y!=None:
      s=x+y
    else:
      s=x
    return s

obj=student()
obj.sum(1,2)
```

```python
class point:
  def __init__(self,x,y):
    self.x=x
    self.y=y

  def __add__(self,other):
    x=self.x+other.x
    y=self.y+other.y
    return point(x,y)

p1= point(1,2)
p2= point(3,4)
print(p1+p2)
```

```python
class point:
  def __init__(self,x,y):
    self.x=x
    self.y=y

  def __add__(self,other):
    x=self.x+other.x
    y=self.y+other.y
    return point(x,y)
  def __str__(self):
    return f"{self.x},{self.y}"

p1= point(1,2)
p2= point(3,4)
print(p1+p2)
```

Overriding:

```python
class college:
  def mymethod(self):
    print('calling parent method')

class school(college):
  def mymethod(self):
    print('calling child method')

obj=school()
obj.mymethod()
```

Super() function:
Python also has a super() function that will make the child class inherit all the methods and properties from its parent.

```python
#parent class:
class FirstClass:
  def __init__(self,v1,v2):
    print('iam in parent class')
    self.v1=v1
    self.v2=v2
  def printValue(self):
    print('values are',self.v1,'and',self.v2)
#second class inheriting from Firstclass
class SecondClass(FirstClass):
  def __init__(self,v1,v2,v3):
    self.v3=v3
    super().__init__(v1,v2)
  def printValue(self):
    print('i am in second class')
  def printValue(self):
    print("iam in second class print")
obj= SecondClass(1,2,3)
obj.printValue()
```
