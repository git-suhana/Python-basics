dictionary:
ordered, changeable and not allow duplicates.
infinte length
iterable
key value pair
value indexed by key

```python
det = {'name':'achu', 'phoneno': '8304', 'job': 'tutor'}
print(det)
```

```python
print(type(det))
```

```python
det['age'] = 43
print(det)
```

added new key:value as age

we can pass list, set as value not key
All data types float, string, int, complex we can pass as key

```python
det = {(3,5): 'sunu', 'phoneno': '8304', 'job': 'tutor'}
print(det)
```

all immutable data can be passed as key such as tuple.


updation:

```python
det = {'name':'lechu', 'phoneno': '0909', 'job': 'tutor'}
det['name'] = "litu"
print(det)
```

```python
det = {'name':'lechu', 'phoneno': '0909', 'job': 'tutor'}
det.update({'job': 'singer'})
det.update({'age':'45'})
print(det)
```
