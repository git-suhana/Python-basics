2. Create a list of characters. Write a program to reverse the order of the elements in the list
without using built-in functions. Use indexing method.

```python

letter=['a','b','c','d','e']
reversed_letter=letter[::-1]
print(reversed_letter)
```

3. Remove empty strings from the list of strings without using any loops
list1 = ["Mike", "", "Emma", "Kelly", "", "Brad"]

```python

List1 = ["Mike", "Emma", "", "Kelly", "Brad"]
List1 = list(filter(None, List1))
print(List1)
```

list1 = [10, 20, [300, 400, [5000, 6000], 500], 30, 40]

```python

list1 = [10, 20, [300, 400, [5000, 6000], 500], 30, 40]
list1[2][2].append(7000)
print(list1)
```

5. You have given a Python list. Write a program to find value 20 in the list, and if it is present,
replace it with 200. Only update the first occurrence of an item.
list1 = [5, 10, 15, 20, 25, 50, 20]
Expected output:
[5, 10, 15, 200, 25, 50, 20]

```python

list1 = [5, 10, 15, 20, 25, 50, 20]
index=list1.index(20)
list1[index]=200
print(list1)
```

6. Create a list of numbers. Write a program to find the range (difference between the largest and
smallest element) of the list.

```python

numbers=[25,15,2,7,9,4,2]
largest=max(numbers)
smallest=min(numbers)
range=largest-smallest
print(range)
```
