Abstraction:
In python, an abstraction is used to hide the irrelevant data.

```python
from abc import ABC,abstractmethod

class Mail(ABC):

  @abstractmethod
  def mileage(self):
    pass

  def fun(self):
    print('fun')

class child(Mail):

  def sum(self):
    print('sum')

  def mileage(self):
    print('mileage')


obj=child()
obj.mileage()
```
