```python
def factorial(num):
  factorial=1
  if num<0:
    print("Factorial does not exist for negative numbers")

  elif num==0:
    print("Factorial of 0 is 1")

  else:

    for i in range(1,num+1):

        factorial= factorial *i
  return factorial
print(factorial(5))
```

Arbitary arguments:
Instead of defining each parameter separately, we can define all the parameters by *args.

```python
def sum1(a,b):
  sum2= a+b
  return sum2

print(sum1(10,20))
```

```python
def sum1(*args):
  print(args)
print(sum1(10,20,30))
```

```python
def sum1(*args):
  print(args)
  sum2 = sum(args)
  return sum2

print(sum1(10,20,30))
```

```python
def func(string):
  vowels= 'aeiouAEIOU'

  for i in string:
    if i in vowels:
      string= string.replace(i,'')
  return string

print(func('hellooo'))
```

```python
def sub(n1,n2):
  if n1<n2:
    (n1,n2)=(n2,n1)

  return n1-n2


def div(n1,n2):
  if n1<n2:
    (n1,n2)=(n2,n1)
  return n1/n2

print(sub(10,20))
print(div(10,20))
```

Decorator function:

```python
def swap_num(fn):
  def wrapper(n1,n2):
    if n1<n2:
      (n1,n2)=(n2,n1)
    return fn(n1,n2)
  return wrapper

@swap_num
def sub(n1,n2):
  return n1-n2

@swap_num
def div(n1,n2):
  return n1/n2

print(sub(10,20))
print(div(10,20))
```
