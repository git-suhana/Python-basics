Conditions:
Used to decide the path of execution.
3 types:
if
else
elif

```python
age = int(input('enter ur age: '))
if age>=60:
  print('eligible for vaccine')
else:
  print('not eligible')
```

```python
num = int(input('enter a number'))
if num>0:
  print('positive')
elif num<0:
  print('negative')
else:
  print('zero')
```

```python
num = int(input('enter a number'))
if num>0:
  print(num, 'is positive')
elif num==0:
  print(num, 'is zero')
else:
  print(num, 'is negative')
```

```python
age = int(input('enter age: '))
if age<10:
  print('lower')
elif age<14:
  print('upper')
elif age<16:
  print('higher')
else:
  print('invalid')
```

```python
age= int(input('enter ur age: '))
health= input('enter ur health status 1 for critical, 2 for normal: ')
if age>60 and health=='1':
  print('not eligible for second dose')
elif age>60 and health=='2':
  print('eligible for second dose')
elif age>18 and health=='1':
  print('not eligible for first dose')
elif age>18 and health=='2':
  print('eligible for first dose')
else:
  print('not eligible')
```
