Filter():
filter function operator on a list and returns a subset of that list after applying the filter rule. The function filter(f,l) needs a function f as it arguments. f returns a boolean value ie either true or false.

```python
list1 = [2,4,5,6,7,8,12,13,15]
x= filter(lambda x:x%2==0,list1)
print(x)
```

```python
print(list(x))
```

```python
list1 = ['raju','ramu','dolu','rema','bolu','arun']
x= filter(lambda x:x.startswith('r'),list1)
print(list(x))
```

map():
it transforms a given list into a new list by transforming each element using a rule, it is used to apply a function to each element of a list or any other irerable.
This is very useful when we do vector mathematics, the map() function will take multiple list operates on the and return the output list, one thing to note is that all the list should be of the same size bcoz map() applies the function to corresponding elements in those lists.


```python
#syntax
#print(list(map(function,list1)))
```

```python
list1= [2,3,4,5]
print(list(map(lambda x:x**2,list1)))
```

```python
list1= ['raju','ramu','dolu','rema','bolu','arun']
list2= list(map(lambda name: name.startswith('r'),list1))
print(list2)
```

zip function:
it take iterables(can be zero or more) makes iterator that aggregates elements based on the iterales passed, and returns an iterator or tuples.
syntax:
zip(*iterables)

```python
rn= [1,2,3]
name= ['raju','ramu','dolu']
data = list(zip(rn,name))
print(data)
```

```python
rn= [1,2,3]
name= ['raju','ramu','dolu']
data= tuple(zip(rn,name))
print(data)
```

```python
rn= [1,2,3]
name= ['raju','ramu','dolu']
data= dict(zip(rn,name))
print(data)
```
