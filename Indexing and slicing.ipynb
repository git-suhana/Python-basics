indexing:
positive indexing and negative indexing.

```python

list = ['rinu','linu','kinu','sini']
print(list[1])
```

```python
print(list[0])
```

```python
print(list[-1])
```

```python
print(list[-3])
```

slicing:
to extract the specific part.

```python
print(list[0:2])
```

```python
print(list[0:])
```

```python
print(list[-4:])
```

```python
print(list[-3:-1])
```

```python
print(list[:4])
```

```python
print(list[:])
```

Append:
to add a value to the list.

```python
list1 = ['rinu','linu','kinu','sini']
print(list1)
list1.append('mini')
print(list1)
```

Extent:
to add nested list

```python
list1= ['rinu','linu','kinu','sini']
list1.extend(['vinu','anu'])
print(list1)
```

Insert:
to insert a value to a given index.


```python
list1.insert(2,'bini')
print(list1)
```

swapping:
to swap two values in a list by swapping the index of values.


```python
list1[0],list1[3]= list1[3],list1[0]
print(list1)
```

Pop():
it removes random item.


```python
list1= ['rinu','linu','kinu','sini']
list1.pop()
print(list1)
```

pop with index value:

```python
list1.pop(1)
print(list1)
```

remove:
directly call the value to be removed.


```python
list1.remove('kinu')
print(list1)
```

delete:
delete the value by giving the index.

```python
del list1[0]
print(list1)
```

sorting:
to sort a list in an order. Default , it sorts in ascending order.

```python
list = [22,32,43,12,14,56]
list.sort()
print(list)
```

to sort in descending order:

```python
list.sort(reverse= True)
print(list)
```

```python
list.reverse()
print(list)
```
