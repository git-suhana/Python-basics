dictionary:
ordered, changeable and not allow duplicates.
infinte length
iterable
key value pair
value indexed by key

```python
det = {'name':'achu', 'phoneno': '8304', 'job': 'tutor'}
print(det)
```

```python
print(type(det))
```

```python
det['age'] = 43
print(det)
```

added new key:value as age

we can pass list, set as value not key
All data types float, string, int, complex we can pass as key

```python
det = {(3,5): 'sunu', 'phoneno': '8304', 'job': 'tutor'}
print(det)
```

all immutable data can be passed as key such as tuple.


updation:

```python
det = {'name':'lechu', 'phoneno': '0909', 'job': 'tutor'}
det['name'] = "litu"
print(det)
```

```python
det = {'name':'lechu', 'phoneno': '0909', 'job': 'tutor'}
det.update({'job': 'singer'})
det.update({'age':'45'})
print(det)
```

Delete:
to remove an item


```python
det = {'name':'lechu', 'phoneno': '0909', 'job': 'tutor'}
del det['name']
print(det)
```

POP()
same as del funcion

```python
det.pop('job')
print(det)
```

popitem():
removes the last added item in the dictionary

```python
det.popitem()
print(det)
```

to find whether the string starts with a specific variable.


```python
x = 'it is good'
x.startswith('i')
```

```python
x.startswith('is',3,5)
```

```python
x.endswith('is',1,5)
```

in endswith function, ending symbol should be exact, starting can be given as one or two before.
